```python
nlp=spacy.load('en')
#r+ reads and w+ overwrites
text=open('../output/CDC_Tweets.txt','r+').read()
#print(text)
```

```python
#turn the string into a corpus for spacy
doc=nlp(text)
len(doc)
#gives the number of words in the file
```

```python
for token in doc:
    print(token)
```

```python
#turn words into list
rows=[]
for token in doc:
    rows.append(token.text)
```

```python
#print(rows)
```

```python
#word list into dataframe
cdc_word_dataframe = pd.DataFrame(rows)
cdc_word_dataframe.columns=['word']
cdc_word_dataframe.head()
```

```python
#count values
cdc_word_dataframe['word'].value_counts()
```

```python
word_count= cdc_word_dataframe.groupby('word').agg({'word':'count'}).rename(columns={'word':'count'})
word_count=word_count.sort_values(by='count', ascending=False)
word_count.head()
```

```python
word_count.to_csv ('../output/cdc_word_count.csv', header=True) 
```
